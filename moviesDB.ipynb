```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/data.csv')
```

```python
del_col  = ['id','imdb_id','popularity','budget_adj','revenue_adj','homepage','keywords','overview', 'production_companies', 'vote_count', 'vote_average']
df = df.drop(del_col,1)
```

```python
df.shape
```

```python
df.head(3)
```

```python
df.drop_duplicates(keep = 'first', inplace = True)
```

```python
row,col = df.shape
```

```python
print(row,col)
```

```python
check_row = ['budget','revenue']
df[check_row] = df[check_row].replace(0,np.NaN)
df.dropna(subset = check_row, inplace = True)
rows,col = df.shape
```

```python
print("After cleaning, we have only {} entries of movies.".format(rows-1))
```

```python
df['runtime'] = df['runtime'].replace(0,np.NaN)
```

```python
df.release_date = pd.to_datetime(df['release_date'])
df.head()
```

```python
df.dtypes
```

```python
change_col_type =  ['budget','revenue']
df[change_col_type] = df[change_col_type].applymap(np.int64)
```

```python
df.dtypes
```

```python
df.rename(columns = {'budget' : 'budget_(in_US-Dollars)', 'revenue' : 'revenue_(in_US-Dollars)'}, inplace = True)
```

```python
df.insert(2, 'profit_(in_US_Dollars)', df['revenue_(in_US-Dollars)'] - df['budget_(in_US-Dollars)'])
```

```python
#*************************QUES/ANS******************************
# which  movie earns the most and least profit

def highest_lowest(column_name):
  highest_id = df[column_name].idxmax()
  highest_details = pd.DataFrame(df.loc[highest_id])

  lowest_id = df[column_name].idxmin()
  lowest_details = pd.DataFrame(df.loc[lowest_id])

  two_in_one_data = pd.concat([highest_details, lowest_details], axis =1)
  return two_in_one_data

highest_lowest('profit_(in_US_Dollars)')
```

```python
highest_lowest('runtime')
```

```python
highest_lowest('budget_(in_US-Dollars)')
```

```python
highest_lowest('revenue_(in_US-Dollars)')
```

```python
#What is the average runtime of all movies?
def average(column_name):
  return df[column_name].mean()
average('runtime')
```

```python
#The average runtime of all movies in this dataset is 109 mins approx.
#We want to get a deeper look and understanding of runtime of all movies
#so Let's plot it.
sns.set_style('darkgrid')
plt.rc('xtick',labelsize=10)
plt.rc('ytick',labelsize=10)

#giving fig size(width,len)
plt.figure(figsize=(9,6), dpi =100)

plt.xlabel('Runtime of movies', fontsize = 15)
plt.ylabel('Number of movies', fontsize = 15)
plt.title('Runtime distrubtion of all movies', fontsize = 18)
plt.hist(df['runtime'], rwidth = 0.9, bins = 31)
plt.show()
```

```python
#In which year we had the most movies making profits?
profit_each_year = df.groupby('release_year')['profit_(in_US_Dollars)'].sum()
plt.figure(figsize=[12,6] , dpi=110)
plt.xlabel('Releaase year of movies', fontsize = 12)
plt.ylabel('Total Profit made by movies', fontsize = 12)
plt.title('Calculating total profits made by all movies in year which it released.')
plt.plot(profit_each_year)
plt.show()
```
